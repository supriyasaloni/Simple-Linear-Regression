# Question 1: 

**What is Simple Linear Regression (SLR)? Explain purpose**


**Answer:**

Simple Linear Regression is a method used to predict one value using another value.

It finds a straight-line relationship between:

One independent variable (X) → the input

One dependent variable (Y) → the output we want to predict



---

**Example (Very Easy)**

You want to predict marks (Y) using study hours (X).

More hours → more marks.

**Simple linear regression draws the best straight line through the data points.**






**Purpose of Simple Linear Regression**

The main purpose of Simple Linear Regression is to understand the relationship between two variables and make predictions.

**1️⃣ Understand the relationship**

It explains how one variable (independent variable X) affects another variable (dependent variable Y).

**2️⃣ Make predictions**

Using the value of X, we can predict the value of Y.
Example:

Hours studied (X) → Exam score (Y)

Work experience (X) → Salary (Y)


**3️⃣ Identify trends**

It helps identify whether the relationship is positive or negative.

Positive: X increases → Y increases

Negative: X increases → Y decreases


**4️⃣ Measure the effect**

The regression equation is:

Y = a + bX

a (intercept) is the value of Y when X = 0


**5️⃣ Support decision-making**

Used in business, economics, healthcare, and research to make data-driven decisions.

 # Question 4:
  **Provide a real-world example where simple linear regression can be applied.**

 
**Answer:**



Simple linear regression is a statistical technique used to study the relationship between one independent variable and one dependent variable. It helps in understanding how a change in one variable affects another and is widely used for prediction and decision-making in real-world situations.


---

**Real-World Example: Prediction of House Price Based on Size**

One of the most common and practical applications of simple linear regression is predicting the price of a house based on its size (area).


---

**Identification of Variables**

**In this example:**
**Independent variable (x):**
Size of the house (in square feet)

**Dependent variable (y):**
Price of the house (in rupees)



---

**Regression Model**

The relationship between house price and house size can be represented using the simple linear regression equation:

$$y = \beta_0 + \beta_1 x + \varepsilon$$

Where:

 $$y= price   of   the   house$$

 $$x= size    of   the   house$$

 $$\beta_0= intercept   (base    price   when   size   is  zero)$$

 $$ \beta_1 x = slope    (change   in   price   for   each   unit   increase   in   size)$$

 $$\varepsilon= error term$$



---

**Explanation of the Relationship**

As the size of the house increases, the price of the house generally increases.
Simple linear regression helps in fitting a straight line that best explains this linear relationship using past data.

The slope  indicates how much the price is expected to increase for every additional square foot of area.


---

**Use of Least Squares Method**

The values of  and  are estimated using the method of least squares, which minimizes the sum of squared differences between the actual house prices and the predicted house prices.

This ensures that the regression line fits the data as accurately as possible.


---

**Practical Importance**

**This model is useful for:**

Real-estate agents to estimate property prices

Buyers to understand fair pricing

Builders and investors for planning and analysis



---



***Advantages of Using Simple Linear Regression***

a. Easy to understand and interpret

b. Useful for prediction

c. Helps in identifying trends and relationships



---



````python
 # Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data (House size in square feet)
X = np.array([500, 800, 1000, 1200, 1500, 1800]).reshape(-1, 1)

# House prices in rupees (in lakhs)
y = np.array([20, 32, 40, 48, 60, 72])

# Create linear regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Predict house price for a new house size
new_size = np.array([[1300]])
predicted_price = model.predict(new_size)

print("Predicted house price for 1300 sq ft:", predicted_price[0], "lakhs")

# Plot the regression line
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("House Size (sq ft)")
plt.ylabel("House Price (lakhs)")
plt.title("House Price Prediction using Simple Linear Regression")
plt.show()
````


**Conclusion**

Simple Linear Regression helps in understanding and predicting real-world problems efficiently. In this example, it is used to predict house prices based on size. As the size of the house increases, the price also increases, showing a positive linear relationship. This method is simple, effective, and widely used in data analysis and decision-making

**What is the method of least squares in linear regression?**


**Answer:**


Method of Least Squares in Linear Regression

The method of least squares is a fundamental statistical technique used in linear regression to estimate the unknown parameters of a regression model. The main objective of this method is to find the best-fitting straight line that represents the relationship between the independent and dependent variables.


---

**Linear Regression Model**

In simple linear regression, the model is expressed as:

$$y = \beta_0 + \beta_1 x + \varepsilon$$


 $$y= price   of   the   house$$

 $$x= size    of   the   house$$

 $$\beta_0= intercept   (base    price   when   size   is  zero)$$

 $$ \beta_1 x = slope    (change   in   price   for   each   unit   increase   in   size)$$

 $$\varepsilon= error term$$


---

**Concept of Errors (Residuals)**

In real-world data, the observed values of  do not lie exactly on a straight line.
The difference between the observed value and the predicted value is called an error or residual.
$$
\text{Residual} = y_i - \hat{y}_i
$$
---

**Idea Behind Least Squares Method**

The method of least squares aims to choose values of  and  such that the sum of the squares of the residuals is minimum.

Mathematically, it minimizes:
$$
\sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$
Squaring the errors ensures that:

Positive and negative errors do not cancel each other

Larger errors are given more importance



---

**Why Squares of Errors Are Used**

1. Squaring removes negative signs


2. Larger deviations are penalized more


3. Mathematical optimization becomes easier




---

**Estimation of Parameters**

Using the least squares principle, we derive formulas to estimate the regression coefficients:

$$\beta_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}$$

$$\beta_0 = \bar{y} - \beta_1 \bar{x}$$

Where:

 $$ \bar{x} = mean           of                  independent                     variable$$

$$\bar{y} = mean of dependent variable$$


These estimates ensure the best possible linear fit to the data.


---

**Graphical Interpretation**

Graphically, the method of least squares finds a straight line such that the vertical distances between the observed data points and the regression line are minimized in a squared sense.


---

**Importance of Least Squares Method**

a.Provides the best linear unbiased estimates

b.Widely used due to its simplicity and effectiveness

c.Forms the foundation of many advanced statistical and machine learning models



---

**Real-World Example**

Suppose we want to predict student marks based on hours studied.
Using the least squares method, we find the regression line that minimizes the difference between actual marks and predicted marks, leading to accurate predictions.


---

**Conclusion**

In conclusion, the method of least squares is a powerful and widely used technique in linear regression. By minimizing the sum of squared errors, it ensures that the regression line fits the data as closely as possible and provides reliable estimates for prediction and analysis.


 # Question 6: 
**What is Logistic Regression? How does it differ from Linear Regression?**


**Answer:**  


Logistic Regression is a supervised machine learning algorithm and a statistical method used mainly for classification problems. It is especially useful when the dependent variable is categorical, most commonly binary (such as 0/1, Yes/No, True/False).

Although it has the word regression in its name, Logistic Regression is actually used for classification, not prediction of continuous values.

Logistic Regression predicts the probability that a given data point belongs to a particular class.
The output value always lies between 0 and 1.

Instead of fitting a straight line like linear regression, logistic regression uses a**logistic (sigmoid)** function to transform the output.


---

**Sigmoid (Logistic) Function**

The sigmoid function converts any real number into a value between 0 and 1.

Mathematical Equation:

$$P(Y = 1) = \frac{1}{1 + e^{-(b_0 + b_1x_1 + b_2x_2 + \dots + b_nx_n)}}$$

Where:

$$ P(Y = 1)= Probability of positive class$$

 $$b_0 = Intercept$$

 $$b_1x_1 + b_2x_2 +.......+ b_nx_n
 = Coefficients$$

$$ x1,x2= Independent variables$$



---

**Working of Logistic Regression**

1. Input features are multiplied with coefficients


2. Linear combination is calculated


3. Sigmoid function is applied


4. Output is converted into probability


5. A threshold value (usually 0.5) is used for classification



**Decision Rule:**

If probability ≥ 0.5 → Class = 1

If probability < 0.5 → Class = 0



---

**Types of Logistic Regression**

**1. Binary Logistic Regression**

Two possible outcomes (Yes/No)



**2. Multinomial Logistic Regression**

More than two classes (No ordering)



**3. Ordinal Logistic Regression**

Ordered classes (Low, Medium, High)





---

 **Difference Between Linear Regression and Logistic Regression**


| Feature | Linear Regression | Logistic Regression |
|------|----------------|----------------|
| Output | Continuous | Probability |
| Use | Prediction | Classification |
| Function | Linear | Sigmoid |
| Range | -∞ to +∞ | 0 to 1 |
| Example | House price | Disease prediction |



---

**Applications of Logistic Regression**

a.Disease prediction (Diabetes: Yes/No)

b.Spam email classification

c.Credit card fraud detection

d.Customer churn prediction

e.Loan approval systems

f.Admission prediction



---

**Advantages of Logistic Regression**

1.Simple and easy to understand

2.Computationally efficient

3.Outputs probability values

4.Works well with small datasets

5.Less prone to overfitting



---

**Limitations of Logistic Regression**

a.Cannot handle complex non-linear relationships

b.Assumes linear relationship between features and log-odds

c.Sensitive to outliers

d.Not suitable for very large feature spaces



---


````python
import numpy as np
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-10, 10, 100)

# Sigmoid (Logistic) function
y = 1 / (1 + np.exp(-x))

# Plot the sigmoid curve
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Sigmoid(X)")
plt.title("Sigmoid Curve in Logistic Regression")
plt.show()
````


 **Conclusion**

Logistic Regression is a powerful and widely used classification algorithm that predicts the probability of an event occurring. Due to its simplicity, interpretability, and efficiency, Logistic Regression is widely used in real-world applications such as healthcare, finance, and marketing

# Question 8: 

**What is the purpose of the R-squared metric in regression analysis?**


**Answer:**

R-squared, also known as the coefficient of determination, is an important statistical measure used in regression analysis to evaluate the performance of a regression model.

**1. Explanation of Variance**
The main purpose of R-squared is to measure how much of the total variation in the dependent variable (Y) is explained by the independent variable(s) (X). It shows the strength of the relationship between variables.

**2. Measure of Goodness of Fit**
R-squared indicates how well the regression line fits the observed data. A higher R² value means the predicted values are closer to the actual values, showing a better model fit.

**3. Range of R-squared**
The value of R-squared lies between 0 and 1:

R² = 0 → The model explains no variation in the dependent variable.

R² = 1 → The model explains 100% of the variation.


**4. Model Evaluation**
R-squared helps analysts and researchers evaluate the accuracy of a regression model and understand whether the independent variables are useful in predicting the dependent variable.

**5. Comparison of Models**
It is useful for comparing multiple regression models built on the same dataset. The model with the higher R-squared value is generally considered better.

**6. Decision-Making Tool**
In business, economics, and data science, R-squared helps in decision making by showing how reliable the predictions of a model are.

**7. Mathematical Expression**

$$R^2 = 1 - \frac{SS_{res}}{SS_{tot}}$$

$$ SS_{res}= Sum of squares of residuals$$

$$ SS_{tot}= Total sum of squares$$


**8. Limitation of R-squared**


R-squared does not indicate causation, and a high R² does not always mean the model is correct. It also increases when more variables are added, even if they are not relevant.



````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 4, 5, 4, 5])

# Regression line
coef = np.polyfit(X.flatten(), Y, 1)
Y_pred = coef[0] * X.flatten() + coef[1]

# R-squared
r2 = r2_score(Y, Y_pred)
print("R-squared value:", r2)

# Plot
plt.figure()
plt.scatter(X, Y)
plt.plot(X, Y_pred)
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (Y)")
plt.title(f"Regression Graph (R² = {r2:.2f})")
plt.show()
````

**Conclusion**


In conclusion, the purpose of R-squared in regression analysis is to measure the explanatory power of a model, assess its accuracy, and help in comparing regression models. It is a vital tool for understanding how well a regression model represents real-world data.

✅ Hence, R-squared is used to evaluate the effectiveness and reliability of regression models.

# Question 9: 
**Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.**

**Answer:**




````python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data (Independent variable X and Dependent variable y)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 6])

# Create Linear Regression model
model = LinearRegression()

# Fit the model
model.fit(X, y)

# Get slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Print results
print("Slope:", slope)
print("Intercept:", intercept)

# Plot the data and regression line
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("X")
plt.ylabel("y")
plt.title("Simple Linear Regression")
plt.show()
````

**Conclusion:**

In this experiment, a Simple Linear Regression model was successfully implemented using the scikit-learn library. The model learned the linear relationship between the independent variable (X) and the dependent variable (y) by estimating the slope and intercept of the best-fit line. The slope indicates how much the dependent variable changes for a one-unit change in the independent variable, while the intercept represents the value of the dependent variable when the independent variable is zero. The plotted regression line closely follows the data points, showing that the model fits the data well. Hence, simple linear regression is an effective and easy-to-interpret method for predicting outcomes and understanding relationships between two variables.

````python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Sample data
X = np.array([1, 2, 3, 4, 5])   # Independent variable
Y = np.array([5, 7, 9, 11, 13]) # Dependent variable

# Calculate slope (β1) and intercept (β0) manually
slope = ( (np.mean(X*Y)) - (np.mean(X)*np.mean(Y)) ) / ( (np.mean(X**2)) - (np.mean(X)**2) )
intercept = np.mean(Y) - slope * np.mean(X)

# Predicted Y values
Y_pred = intercept + slope * X

# Plot
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.title('Simple Linear Regression Graph')
plt.legend()
plt.grid(True)
plt.show()
````
